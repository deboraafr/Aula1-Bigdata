As 10 marcas de carros líderes em faturamento

```python
import pandas as pd
import numpy as np
```

```python
marcas = ["Fiat", "Toyota", "Chevrolet", "Volkswagen", "Jeep", "Hyundai", "Renault", "Nissan", "Honda", "Caoa Chery"]
marcas
```

```python
bilhões = [45.3, 39, 37.7, 31.4, 26.6, 19.7, 12.5, 8.2, 7.5, 6.2]
```

```python
serie1 = pd.Series (bilhões, marcas)
serie1
```

```python
import matplotlib.pyplot as plt
```

```python
serie1.plot()
```

De acordo com o gráfico produzido, é possível perceber o faturamento em bilhões de cada uma das 10 marcas com os maiores faturamentos no ramo de automóveis. Fica notório também a gritante diferença entre o faturamento das marcas que ocupam as primeiras posições e as que ocupam as últimas posições, mesmo sendo todas elas consideradas líderes em faturamento.
